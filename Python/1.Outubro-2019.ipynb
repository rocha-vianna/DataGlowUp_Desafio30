# Arrumando o CSV de Outubro/2019

```python
# Importando os pacotes necessÃ¡rios
import pandas as pd

# Lendo o arquivo .CSV
df_oct = pd.read_csv('source/1.2019-Oct.csv')
```

```python
# Visualizando as 5 primeiras linhas do 'df_oct'
df_oct.head()
```

```python
# Dividindo a coluna 'event_time' em 3 novas colunas
df_oct[['Data', 'Hora', 'Zona']] = df_oct['event_time'].str.split(' ', expand=True)
```

```python
# Excluindo as colunas 'event_time' e 'Zona'
df_oct = df_oct.drop(columns=['event_time', 'Zona'])

# Visualizando as 5 primeiras linhas do 'df_oct'
df_oct.head()
```

```python
# Reorganizando as colunas
df_oct = df_oct[['Data', 'Hora', 'event_type', 'product_id', 'category_id', 'category_code', 'brand', 'price', 'user_id', 'user_session']]

# Visualizando as 5 primeiras linhas do 'df_oct'
df_oct.head()
```

```python
# Salvando o df num arquivo .CSV
df_oct.to_csv('2019Oct_SQL.csv', index=False)
```
