# Arrumando o CSV de Dezembro/2019

```python
# Importando os pacotes necess√°rios
import pandas as pd

# Lendo o arquivo .CSV
df_dec = pd.read_csv('source/3.2019-Dec.csv')
```

```python
# Visualizando as 5 primeiras linhas do 'df_dec'
df_dec.head()
```

```python
# Dividindo a coluna 'event_time' em 3 novas colunas
df_dec[['Data', 'Hora', 'Zona']] = df_dec['event_time'].str.split(' ', expand=True)
```

```python
# Excluindo as colunas 'event_time' e 'Zona'
df_dec = df_dec.drop(columns=['event_time', 'Zona'])

# Visualizando as 5 primeiras linhas do 'df_dec'
df_dec.head()
```

```python
# Reorganizando as colunas
df_dec = df_dec[['Data', 'Hora', 'event_type', 'product_id', 'category_id', 'category_code', 'brand', 'price', 'user_id', 'user_session']]

# Visualizando as 5 primeiras linhas do 'df_dec'
df_dec.head()
```

```python
# Salvando o df num arquivo .CSV
df_dec.to_csv('2019Dec_SQL.csv', index=False)
```
