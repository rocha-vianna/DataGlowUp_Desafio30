# Arrumando o CSV de Janeiro/2020

```python
# Importando os pacotes necess√°rios
import pandas as pd

# Lendo o arquivo .CSV
df_jan = pd.read_csv('source/4.2020-Jan.csv')
```

```python
# Visualizando as 5 primeiras linhas do 'df_jan'
df_jan.head()
```

```python
# Dividindo a coluna 'event_time' em 3 novas colunas
df_jan[['Data', 'Hora', 'Zona']] = df_jan['event_time'].str.split(' ', expand=True)
```

```python
# Excluindo as colunas 'event_time' e 'Zona'
df_jan = df_jan.drop(columns=['event_time', 'Zona'])

# Visualizando as 5 primeiras linhas do 'df_jan'
df_jan.head()
```

```python
# Reorganizando as colunas
df_jan = df_jan[['Data', 'Hora', 'event_type', 'product_id', 'category_id', 'category_code', 'brand', 'price', 'user_id', 'user_session']]

# Visualizando as 5 primeiras linhas do 'df_jan'
df_jan.head()
```

```python
# Salvando o df num arquivo .CSV
df_jan.to_csv('2020Jan_SQL.csv', index=False)
```
