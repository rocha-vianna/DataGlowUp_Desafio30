# Arrumando o CSV de Novembro/2019

```python
# Importando os pacotes necess√°rios
import pandas as pd

# Lendo o arquivo .CSV
df_nov = pd.read_csv('source/2.2019-Nov.csv')
```

```python
# Visualizando as 5 primeiras linhas do 'df_nov'
df_nov.head()
```

```python
# Dividindo a coluna 'event_time' em 3 novas colunas
df_nov[['Data', 'Hora', 'Zona']] = df_nov['event_time'].str.split(' ', expand=True)
```

```python
# Excluindo as colunas 'event_time' e 'Zona'
df_nov = df_nov.drop(columns=['event_time', 'Zona'])

# Visualizando as 5 primeiras linhas do 'df_nov'
df_nov.head()
```

```python
# Reorganizando as colunas
df_nov = df_nov[['Data', 'Hora', 'event_type', 'product_id', 'category_id', 'category_code', 'brand', 'price', 'user_id', 'user_session']]

# Visualizando as 5 primeiras linhas do 'df_nov'
df_nov.head()
```

```python
# Salvando o df num arquivo .CSV
df_nov.to_csv('2019Nov_SQL.csv', index=False)
```
