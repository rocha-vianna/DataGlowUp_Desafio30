# Arrumando o CSV de Fevereiro/2020

```python
# Importando os pacotes necessários
import pandas as pd

# Lendo o arquivo .CSV
df_feb = pd.read_csv('source/5.2020-Feb.csv')
```

```python
# Visualizando as 5 primeiras linhas do 'df_oct'
df_feb.head()
```

```python
# Dividindo a coluna 'event_time' em 3 novas colunas
df_feb[['Data', 'Hora', 'Zona']] = df_feb['event_time'].str.split(' ', expand=True)
```

```python
# Excluindo as colunas 'event_time' e 'Zona'
df_feb = df_feb.drop(columns=['event_time', 'Zona'])

# Visualizando as 5 primeiras linhas do 'df_oct'
df_feb.head()
```

```python
# Reorganizando as colunas
df_feb = df_feb[['Data', 'Hora', 'event_type', 'product_id', 'category_id', 'category_code', 'brand', 'price', 'user_id', 'user_session']]

# Visualizando as 5 primeiras linhas do 'df_oct'
df_feb.head()
```

```python
# Salvando o df num arquivo .CSV
df_feb.to_csv('2020Feb_SQL.csv', index=False)
```
